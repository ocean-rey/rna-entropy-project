```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from pathlib import Path
```

```python
def get_distribution_jumping(seq, k_mer_dict = {}, window = 3):
    length = len(seq)-1
    window_open = 0
    window_close = 3
    while(window_close<=length):
        k_mer = seq[window_open:window_close]
        if k_mer in k_mer_dict.keys():
            k_mer_dict[k_mer] += 1
        else:
            k_mer_dict[k_mer] = 1
        window_open += 3
        window_close +=3
    return k_mer_dict
```

```python
def get_distribution_sliding(seq, k_mer_dict = {}):
    length = len(seq)-1
    window_open = 0
    window_close = 3
    while(window_close<=length):
        k_mer = seq[window_open:window_close]
        if k_mer in k_mer_dict.keys():
            k_mer_dict[k_mer] += 1
        else:
            k_mer_dict[k_mer] = 1
        window_open += 1
        window_close +=1
    return k_mer_dict
```

```python
rona = Path('covid/sequence.fasta').read_text()
rona = rona.replace('\n','')
```

```python
txt = Path('covid/Genes.GFF3').read_text()
rona_map = {}
txt = txt.splitlines()
for line in txt[3:]:
    line = line.split("\t")
    annot = line[2]
    if annot in rona_map.keys():
        rona_map[annot].append(rona[int(line[3])-1:int(line[4])])
    else:
        rona_map[annot] = [rona[int(line[3])-1:int(line[4])]]
```

```python
full_dist = get_distribution_jumping(rona)
full_dist = sorted(full_dist.items(), key=lambda kv: kv[1], reverse=False)
x_full, y_full = zip(*full_dist)
plt.bar(x_full, y_full)
plt.show()
(svals, ppcc) = stats.ppcc_plot(x=y_full, a=y_full[0],
                                b=y_full[-1],plot=plt,
                                dist="pareto")
plt.show()
full_avg_cc = ppcc.mean()
print("Average Correlation Coefficient:{}".format(full_avg_cc))
```

```python
pareto_ppcc_map = {}
for annotation in rona_map.keys():
    dist = {}
    for region in rona_map.get(annotation):
        dist = get_distribution_jumping(region, dist)
    dist = sorted(dist.items(), key=lambda kv:  kv[1], reverse=False)
    x, y = zip(*dist)
    plt.bar(x, y)
    plt.title("{} k_mer distribution".format(annotation))
    plt.show()
    (svals, ppcc) = stats.ppcc_plot(x=y, a=y[0],
                                    b=y[-1],plot=plt,
                                    dist="pareto")
    pareto_ppcc_map[annotation] = ppcc.mean()
    plt.title("{} pareto ppcc plot. mean pcc = {}".format(annotation, ppcc.mean()))
    plt.show()
    stats.probplot(y, dist="pareto", plot=plt, sparams = [y[0],0])
    plt.title("{} Q-Q plot (pareto distribution)".format(annotation))
    plt.show()
```

```python
tukeylambda_ppcc_map = {}
for annotation in rona_map.keys():
    dist = {}
    for region in rona_map.get(annotation):
        dist = get_distribution_jumping(region, dist)
    dist = sorted(dist.items(), key=lambda kv:  kv[1], reverse=False)
    x, y = zip(*dist)
    plt.plot(x, y)
    plt.title("{} trimer distribution".format(annotation))
    plt.show()
    (svals, ppcc) = stats.ppcc_plot(x=y, a=y[0],
                                    b=y[-1],plot=plt)
    tukeylambda_ppcc_map[annotation] = ppcc.mean()
    plt.title("{} ppcc plot. mean pcc = {}".format(annotation, ppcc.mean()))
    plt.show()
    stats.probplot(y, dist="norm", plot=plt)
    plt.title("{} Q-Q plot".format(annotation))
    plt.show()
    
```
